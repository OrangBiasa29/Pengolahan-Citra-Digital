# Baca Gambar

```python
img_asli = Image.open("gambar.jpg")
img_asli.save("gambar.jpeg", "jpeg")
img = cv2.imread("gambar.jpeg")

# mengubah format image yang awalnya BGR (bawaan dari OpenCV) menjadi RGB
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.title("Gambar Asli")
plt.axis("off")
plt.show()
```

# Mean Filter, Gaussian Filter, Median Filter, dan Sharpening Filter dengan High Resolution Image

```python
# 1. Mean Filter (Averaging)
mean_filter = cv2.blur(img_rgb, (5,5))

# 2. Gaussian Filter
gaussian_filter = cv2.GaussianBlur(img_rgb, (5,5), 0)

# 3. Median Filter
median_filter = cv2.medianBlur(img_rgb, 5)

# 4. Sharpening Filter
kernel_sharpening = np.array([[0,-1,0],
                              [-1,5,-1],
                              [0,-1,0]])
sharpened = cv2.filter2D(img_rgb, -1, kernel_sharpening)

# 5. Laplacian Filter (edge enhancement)
gray = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2GRAY)
laplacian = cv2.Laplacian(gray, cv2.CV_64F)
laplacian = cv2.convertScaleAbs(laplacian)

# Tampilkan hasil
titles = ["Original", "Mean Filter", "Gaussian Filter",
          "Median Filter", "Sharpened", "Laplacian"]
images = [img_rgb, mean_filter, gaussian_filter,
          median_filter, sharpened, laplacian]

plt.figure(figsize=(15,8))
for i in range(len(images)):
    plt.subplot(2,3,i+1)
    if len(images[i].shape) == 2:  # grayscale
        plt.imshow(images[i], cmap='gray')
    else:
        plt.imshow(images[i])
    plt.title(titles[i])
    plt.axis("off")
plt.tight_layout()
plt.show()
```

## Analisis
## Mean Filter

* Efek smoothing terlihat jelas.

* Gambar menjadi blur karena mean filter menghitung rata-rata piksel tetangga.

* Akibatnya detail halus hilang (misalnya kerutan wajah dan tahi lalat di samping matanya).

* Cocok dipakai untuk mengurangi noise, tetapi kurang bagus untuk mempertahankan ketajaman.

## Gaussian Filter

* Sama-sama smoothing,seperti mean, tetapi hasilnya lebih natural dibanding mean filter.

* Blur lebih halus dan menyebar, transisi warna lebih lembut.

* Masih menghilangkan detail, tapi tidak sekeras mean filter (tahi lalat masih terlihat).

* Banyak dipakai untuk mengurangi noise random (Gaussian noise).

## Median Filter

* Juga menghasilkan smoothing, tapi lebih efektif untuk menghapus noise "salt-and-pepper" (bintik hitam/putih).

* Pada gambar ini, wajah & tangan lebih mulus dibanding Gaussian/Mean, namun garis besar masih terlihat.

* Median lebih preserve edges (tepi tidak terlalu kabur).

## Sharpened

* Detail citra ditingkatkan.

* Garis wajah, jas, podium, dan tangan terlihat lebih jelas.

* Efek sharpening menambahkan kontras lokal di sekitar tepi (edges).

* Cocok dipakai untuk memperjelas gambar blur.

## Laplacian

* Menunjukkan edge map (peta tepi) dari gambar.

* Semua area dengan perubahan intensitas tajam (batas wajah, tangan, podium, jas) terlihat jelas.

* Laplacian tidak menghasilkan gambar “indah”, tetapi dipakai untuk deteksi tepi atau digabung dengan gambar asli untuk enhancement.

## Kesimpulan

* Mean & Gaussian → memperhalus, tapi menghilangkan detail (gambar jadi blur).

* Median → memperhalus tapi tetap menjaga tepi → ideal untuk noise impulsif.

* Sharpening → menegaskan detail → membuat gambar lebih jelas.

* Laplacian → mengekstrak tepi → dasar untuk edge enhancement atau deteksi objek.

# Dark Image Enhancement

```python
gamma = 0.5
gamma_corrected = np.array(255*( (img_rgb/255.0) ** (1/gamma) ), dtype='uint8')

plt.imshow(gamma_corrected)
plt.title("Dark Image Enhanced")
plt.axis("off")
plt.show()
```

Gambar lebih gelap.

# Bright Image Enhancement

```python
gamma = 2
gamma_corrected = np.array(255*( (img_rgb/255.0) ** (1/gamma) ), dtype='uint8')

plt.imshow(gamma_corrected)
plt.title("Bright Image Enhanced")
plt.axis("off")
plt.show()
```

Gambar lebih terang.

# Bright Image Reduce Enhancement



```python
alpha = 0.8   # contrast control
beta = -30    # brightness control (nilai negatif untuk mengurangi kecerahan)
bright_reduced = cv2.convertScaleAbs(img_rgb, alpha=alpha, beta=beta)

plt.imshow(bright_reduced)
plt.title("Bright Image Reduced (Brightness Adjustment)")
plt.axis("off")
plt.show()
```

1. Kecerahan berkurang
    * Wajah tetap terlihat jelas, tapi cahaya berlebih sudah berkurang.

2. Detail lebih terjaga

    * Area jas, podium, dan background masih memiliki tekstur.

    * Tidak terjadi over darkening (tidak kehilangan detail di area gelap).

3. Warna lebih natural

    * Kulit wajah terlihat lebih realistis dibanding versi terlalu terang.

    * Background kayu jadi lebih pekat, sehingga subjek (pembicara) lebih menonjol.

4. Keseimbangan lebih baik
    * Hasilnya lebih seimbang dibanding pengurangan brightness ekstrem.

# Low Contrast Image Enhancement

```python
lab = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)
l, a, b = cv2.split(lab)

clahe = cv2.createCLAHE(clipLimit=3.0, tileGridSize=(8,8))
cl = clahe.apply(l)
limg = cv2.merge((cl,a,b))

contrast_enhanced = cv2.cvtColor(limg, cv2.COLOR_LAB2RGB)

plt.imshow(contrast_enhanced)
plt.title("Low-Contrast Image Enhanced (CLAHE)")
plt.axis("off")
plt.show()
```

1. Kontras lebih tinggi
    * Wajah, jas, dan podium terlihat lebih menonjol.

    * Area gelap dan terang jadi lebih terpisah.

2. Detail lebih jelas

    * Garis wajah, rambut, dan lipatan jas muncul lebih tajam dibanding versi low-contrast sebelumnya.

    * Logo podium lebih mudah dibedakan dari latar belakang.

3. Warna tampak lebih hidup

    * Karena kontras meningkat, warna baju, kulit, dan background kayu terlihat lebih natural (tidak pucat).
